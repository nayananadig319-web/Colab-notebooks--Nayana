```python
crops=["wheat","rice","maize"]
yield_per_acre=[20,30,25]
area=10

for i in range(len(crops)):
  total_yield =yield_per_acre[i]*area
  print(crops[i],"total yield=",total_yield,"quintals")
```

```python
rainfall=(120,80,100,150,200,90)
highest=max(rainfall)
lowest=min(rainfall)
avearge=sum(rainfall)/len(rainfall)

print("Highest rainfall=",highest,"mm")
print("lowest rainfall=",lowest,"mm")
print("Average rainfall=",round(avearge,2),"mm")
```

```python
crop_price = {"Wheat": 2400, "Rice": 3000, "Maize": 2100}
print("Crop Prices:")

for crop, price in crop_price.items():
    print(crop, ":",price)
highest_crop = max(crop_price, key=crop_price.get)
print("Highest priced crop =", highest_crop)

new_crop = input("Enter new crop name: ")
new_price = float(input("Enter price per quintal: "))

crop_price[new_crop] = new_price

print("\nUpdated Crop Price List:")
for crop, price in crop_price.items():
    print(crop, ":", price)
```

```python
yield_amount = float(input("Enter total yield (in quintals): "))
price = float(input("Enter market price per quintal: "))
cost = float(input("Enter total cost of cultivation: "))

profit = (yield_amount * price) - cost

if profit > 0:
    print("Profit =", profit)
elif profit < 0:
    print("Loss =",abs(profit))
else:
    print("Break-even (No Profit, No Loss)")
```
