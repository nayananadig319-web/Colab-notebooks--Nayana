```python
rainfall=[]
for i in range(5):
  rainfall_value=float(input("enter rainfall for day:"))
  rainfall.append(rainfall_value+1)
  total=0
  for r in rainfall:
    total=total+r
    average=total/5
    print("average rainfall:",average)
    if average>=100:
      print("sufficient rainfall")
    else:
      print("insufficient rainfall")
```
