```python
import pandas as pd
import numpy as np
# Dataset
data={
    'Weather':
    ['sunny','sunny','cloudy','rainy','rainy','rainy','cloudy','sunny','sunny','rainy','sunny','cloudy','cloudy','rainy'],
    'Temperature':
    ['hot','hot','hot','mild','cool','cool','cool','mild','cool','mild','mild','mild','hot','mild'],
    'Soil moisture':
    ['dry','dry','dry','dry','moist','moist','moist','dry','moist','moist','moist','dry','moist','dry'],
    'Wind':
    ['weak','strong','weak','weak','weak','strong','strong','weak','weak','weak','strong','strong','weak','strong'],
    'Irrigate':
    ['no','no','yes','yes','yes','no','yes','no','yes','yes','yes','yes','yes','no']

}

df=pd.DataFrame(data)

# Step 1: ENTROPY

def entropy(target):
  values,counts=np.unique(target,return_counts=True)
  prob=counts/counts.sum()
  return -np.sum(prob*np.log2(prob))

# Step 2: INFORMATION GAIN

def information_gain(data,feature,target_name):
  # Corrected function name from 'informmation_gain' to 'information_gain'
  # Corrected typo from 'targget_name' to 'target_name'
  total_entropy=entropy(data[target_name])

  values,counts=np.unique(data[feature],return_counts=True)
  weighted_entropy=0

  for i in range(len(values)):
    subset=data[data[feature]==values[i]]
    weighted_entropy+=(counts[i]/np.sum(counts))*entropy(subset[target_name])
  return total_entropy-weighted_entropy # Moved return statement outside the loop

# step 3:ID3 ALGORITHM

def id3(data,features,target_name,depth=0):
  indent="  "*depth # Consistent indentation for each level

  # Base Case 1: If all target values in the current subset are the same
  if len(np.unique(data[target_name]))==1:
    return np.unique(data[target_name])[0]

  # Base Case 2: If there are no features left to split on
  if len(features)==0:
    return data[target_name].mode()[0]

  gains=[]
  for feature in features:
    gain=information_gain(data,feature,target_name) # Correct function name
    gains.append(gain)

  # Find the feature with the highest information gain
  best_feature_index = np.argmax(gains)
  best_feature = features[best_feature_index] # Define best_feature before use

  # Moved this print statement from outside the function and placed it here
  print(f"\n{indent}Evaluating features at depth {depth}:") # Added formatting
  print(f"{indent}Best Feature: {best_feature} (Gain: {gains[best_feature_index]:.4f})") # Added gain value

  tree={best_feature:{}}

  # SPLIT DATASET
  for value in np.unique(data[best_feature]):
    print(f"{indent}  Splitting {best_feature}={value}") # Added more indentation for visual clarity
    subset=data[data[best_feature]==value] # Corrected typo `data_feature` to `best_feature`

    if subset.shape[0]==0:
      # If subset is empty, assign the mode of the parent's target
      tree[best_feature][value]=data[target_name].mode()[0]
    else:
      # Recursively call id3 for the subset
      remaining_features=[f for f in features if f!=best_feature]
      subtree=id3(subset,remaining_features,target_name,depth+1)
      tree[best_feature][value]=subtree
  return tree # Correctly indented inside the function

# Step 4: PRINT TREE FUNCTION

def print_tree(tree,indent=""):
  if not isinstance(tree,dict):
    print(f"{indent}--> {tree}") # Corrected print statement for better formatting
    return

  for feature,branches in tree.items():
    print(f"{indent}{feature}:") # Added colon for better readability
    for value,subtree in branches.items():
      print(f"{indent}  {value}:") # Consistent indentation and colon
      print_tree(subtree,indent+"    ") # Increased indentation for deeper levels

# Step5: BUILD TREE

features=list(df.columns[:-1]) # Corrected variable name from 'feature' to 'features'
tree_model=id3(df,features,'Irrigate')

# Step6:DISPLAY TREE

print("\n Final Decision Tree:\n")
print_tree(tree_model)
```
