```python
#Step 1: import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

# Step2: Create dataset
data={
    'Weather':['Sunny','Sunny','Rainy','Rainy','Cloudy','Sunny','Cloudy','Rainy'],
    'Soil':['Dry','Wet','Wet','Dry','Dry','Dry','Wet','Wet'],
    'Temperature':['Hot','Hot','Cool','Cool','Cool','Cool','Hot','Hot'],
    'Irrigate':['Yes','No','No','Yes','Yes','Yes','No','No']
}
df=pd.DataFrame(data)
print("Dataset:\n",df)

# Step 3: Encode categorical features
le_weather=LabelEncoder()
le_soil=LabelEncoder()
le_temp=LabelEncoder()
le_target=LabelEncoder()

df['Weather']=le_weather.fit_transform(df['Weather'])
df['Soil']=le_soil.fit_transform(df['Soil'])
df['Temperature']=le_temp.fit_transform(df['Temperature'])
df['Irrigate']=le_target.fit_transform(df['Irrigate'])

print("\n Encoded Dataset:\n",df)

# Step4:Split features and target
X=df[['Weather','Soil','Temperature']]
Y=df['Irrigate']

# Step 5: Train naive bayes model
model=CategoricalNB()
model.fit(X,Y)

# Step 6: Test input(Sunny,Dry,Cool)

test_data=pd.DataFrame({
    'Weather':['Sunny'],
    'Soil':['Dry'],
    'Temperature':['Cool']
})

# Encode test data using same encoders
test_data['Weather']=le_weather.transform(test_data['Weather'])
test_data['Soil']=le_soil.transform(test_data['Soil'])
test_data['Temperature']=le_temp.transform(test_data['Temperature'])

# step7: Prediction

prediction=model.predict(test_data)

# convert back to original label
result=le_target.inverse_transform(prediction)
print("\n Test Input:Sunny,Dry,Cool")
print("Prediction:",result[0])

# Step 8: Probability output(optional)
prob=model.predict_proba(test_data)
print("Class Probabilities:",prob)
```

```python
#Step 1: import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

# Step2: Create dataset
data={
    'Weather':['Sunny','Sunny','Rainy','Rainy','Cloudy','Cloudy','Sunny','Rainy'],
    'Humidity':['High','Low','HIgh','High','Low','High','High','Low'],
    'CropStage':['Flowering','Vegetative','Flowering','Vegetative','Vegetative','Flowering','Vegetative','Flowering'],
    'PestAttack':['Yes','No','Yes','Yes','No','Yes','No','No']
}
df=pd.DataFrame(data)
print("Dataset:\n",df)

# Step 3: Encode categorical features
le_weather=LabelEncoder()
le_humidity=LabelEncoder()
le_cropstage=LabelEncoder()
le_pestattack=LabelEncoder()

df['Weather']=le_weather.fit_transform(df['Weather'])
df['Humidity']=le_humidity.fit_transform(df['Humidity'])
df['CropStage']=le_cropstage.fit_transform(df['CropStage'])
df['PestAttack']=le_pestattack.fit_transform(df['PestAttack'])

print("\n Encoded Dataset:\n",df)

# Step4:Split features and target
X=df[['Weather','Humidity','CropStage']]
Y=df['PestAttack']

# Step 5: Train naive bayes model
model=CategoricalNB()
model.fit(X,Y)

# Step 6: Test input(Sunny,Dry,Cool)

test_data=pd.DataFrame({
    'Weather':['Sunny'],
    'Humidity':['High'],
    'CropStage':['Flowering']
})

# Encode test data using same encoders
test_data['Weather']=le_weather.transform(test_data['Weather'])
test_data['Humidity']=le_humidity.transform(test_data['Humidity'])
test_data['CropStage']=le_cropstage.transform(test_data['CropStage'])

# step7: Prediction

prediction=model.predict(test_data)

# convert back to original label
result=le_target.inverse_transform(prediction)
print("\n Test Input:Sunny,High,Flowering")
print("Prediction:",result[0])

# Step 8: Probability output(optional)
prob=model.predict_proba(test_data)
print("Class Probabilities:",prob)
```
