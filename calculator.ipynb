```python
def add(x,y):
  return x+y
def subtract(x,y):
  return x-y
def multiply(x,y):
  return x*y
def divide(x,y):
  if y==0:
    return "ERROR!DIVISION BY ZERO."
  else:
    return x/y
def get_input():
  try:
    x=float(input("Enter the first number: "))
    y=float(input("Enter the second number: "))
    return x,y
  except ValueError: # Corrected typo: Valueerror to ValueError
    print("Invalid input!Please enter valid numbers.")
    return get_input()
def print_menu():
  print("\nSelect the operation: ")
  print("1.Add")
  print("2.Subtract")
  print("3.Multiply")
  print("4.Divide")
  print("5.Exit")
def calculator():
  while True:
    print_menu()
    choice=input("Enter choice(1/2/3/4/5): ") # Moved inside the loop
    if choice=='5':
      print("Exiting the calculator.")
      break # Now correctly inside the while loop

    x,y=get_input()
    if choice=='1':
      print(f"{x}+{y}={add(x,y)}")
    elif choice=='2':
      print(f"{x}-{y}={subtract(x,y)}") # Corrected: was add(x,y)
    elif choice=='3':
      print(f"{x}*{y}={multiply(x,y)}") # Corrected: was add(x,y)
    elif choice=='4':
      print(f"{x}/{y}={divide(x,y)}") # Corrected: was add(x,y)
    else:
      print("Invalid input!Please select a valid operation(1/2/3/4/5): ")
calculator()
print("DONE WITH THE PROGRAM!")
```
