```python
def load_data(filename):
  """reads crop data from a file and returns a list of dictionaries"""
  crops=[]
  file=open(filename,"r")
  for line in file:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "rainfall":float(data[3]),
          "fertilizer":float(data[4]),
          "yield":float(data[5]),
      }
      crops.append(crop)
  file.close()
  return crops

def calculate_yield_per_area(crop):
  """returns yield per unit area"""
  return crop["yield"]/crop["area"]

def display_crops(crops):
  """displays all crop record"""
  print("\ncrop records:")
  for crop in crops:
    ypa=calculate_yield_per_area(crop)
    print(crop["name"],"-",crop["season"],"-yield/area:",round(ypa,2))

def find_highest_and_lowest_yield(crops):
  """finds crops with highest and lowest yield per area"""
  highest=crops[0]
  lowest=crops[0]

  for crop in crops:
    if calculate_yield_per_area(crop)>calculate_yield_per_area(highest):
      highest=crop
    if calculate_yield_per_area(crop)<calculate_yield_per_area(lowest):
      lowest=crop
  return highest,lowest

def seasonal_summary(crops):
  """generates seasonal yield summary using dictionary"""
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculate_yield_per_area(crop)
    if season not in summary:
      summary[season]=[]
    summary[season].append(ypa)
  return summary

def recommend_crops(crops, season, max_rainfall, max_fertilizer):
    print("Recommended crops:")

    for crop in crops:
        if (crop["season"] == season and
            crop["rainfall"] <= max_rainfall and
            crop["fertilizer"] <= max_fertilizer):
            print(crop["name"])


def crops_in_multiple_seasons(crops):
  crop_seasons={}
  for crop in crops:
    name=crop["name"]
    season=crop["season"]
    if name not in crop_seasons:
            crop_seasons[name] = []
    if season not in crop_seasons[name]:
            crop_seasons[name].append(season)

  for name, seasons in crop_seasons.items():
        if len(seasons) > 1:
            print(name, "grown in seasons:", seasons)


def best_crop_per_season(crops):
    best = {}

    for crop in crops:
        season = crop["season"]

        if season not in best:
            best[season] = crop
        else:
            if calculate_yield_per_area(crop) > calculate_yield_per_area(best[season]):
                best[season] = crop

    for season, crop in best.items():
        print(season, ":", crop["name"])

def rank_crops_by_yield(crops):
    ranked = sorted(crops, key= calculate_yield_per_area, reverse=True)

    rank = 1
    for crop in ranked:
        print(rank, crop["name"], "-", round(calculate_yield_per_area(crop), 2))
        rank += 1


def rainfall_efficiency_analysis(crops, max_rainfall):
    print("Crops with rainfall <=", max_rainfall)

    for crop in crops:
        if crop["rainfall"] <= max_rainfall:
            print(crop["name"], "-", crop["yield"])


def find_max_rainfall(crops):
    max_rain = crops[0]["rainfall"]

    for crop in crops:
        if crop["rainfall"] > max_rain:
            max_rain = crop["rainfall"]
    return max_rain


def fertilizer_efficiency(crop):
    return crop["yield"] / crop["fertilizer"]

def most_fertilizer_efficient_crop(crops):
    best = crops[0]

    for crop in crops:
        if fertilizer_efficiency(crop) > fertilizer_efficiency(best):
            best = crop

    print("Most Fertilizer Efficient Crop:", best["name"])


def crop_count_per_season(crops):
    count = {}

    for crop in crops:
      season=crop["season"]
      if season not in count:
        count[season]=0
      count[season]+=1
    print("\n crop count per season")
    for season,value in count.items():
      print(season,":",value)


def write_report(filename,crops,highest,lowest,summary):
  """writes analysis results to output file"""
  file=open(filename,"w")
  file.write("agricultural data analysis report\n\n")
  file.write("crop yield per unit area:\n")

  for crop in crops:
    ypa=calculate_yield_per_area(crop)
    file.write(crop["name"]+":"+str(round(ypa,2))+"\n")
  file.write("\n highest yield crop:\n")
  file.write(highest["name"]+"\n")
  file.write("\n lowest yield crop:\n")
  file.write(lowest["name"]+"\n")
  file.write("\n seasonal average yield:\n")

  for season in summary:
    avg=sum(summary[season])/len(summary[season])
    file.write(season + ":" + str(round(avg,2))+"\n")
  file.close()

#------main program------

def main():
  input_file="crops data2.txt"
  output_file="analysis report.txt"
  crops=load_data(input_file)

  for crop in crops:
    print(crop)
  for crop in crops:
    for key,value in crop.items():
      print(f"{key}:{value}")
    print("-----")
  display_crops(crops)

  highest,lowest=find_highest_and_lowest_yield(crops)
  print("\n highest:",highest)
  print("\n lowest:",lowest)
  print("\n")

  summary=seasonal_summary(crops)
  print("seasonal summary \n",summary)
  print("\n")
  crop_count_per_season(crops)
  print("\n")

  most_fertilizer_efficient_crop(crops)
  print("\n")

  max_rainfall = find_max_rainfall(crops)
  rainfall_efficiency_analysis(crops, max_rainfall)
  print("\n")
  rank_crops_by_yield(crops)
  print("\n")
  best_crop_per_season(crops)
  print("\n")
  crops_in_multiple_seasons(crops)
  print("\n")

  recommendation_season = "Rabi"
  recommendation_max_rainfall = 1200
  recommendation_max_fertilizer = 200
  print("\n")

  recommend_crops(crops, recommendation_season, recommendation_max_rainfall, recommendation_max_fertilizer)


  write_report(output_file, crops, highest, lowest, summary)
  print("\n analysis complete.report written to",output_file)

main()
```
