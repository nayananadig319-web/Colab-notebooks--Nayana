```python
print("=====FARMERS INPUT BASED PROFIT ANALYSIS=====")

seed_cost=float(input("enter seed cost(INR):"))
fertilizer_cost=float(input("enter fertilizer cost(INR):"))
pesticide_cost=float(input("enter pesticide cost(INR):"))
labor_cost=float(input("enter pesticide cost(INR):"))
irrigation_cost=float(input("enter labor cost(INR):"))
total_input_cost=seed_cost+fertilizer_cost+pesticide_cost+labor_cost+irrigation_cost

print("\n TOTAL INPUT COST(INR):",total_input_cost)

total_yield_kg=float(input("\n enter total crop yield(kg):"))
price_per_kg=float(input("enter selling price per kg(INR):"))
total_revenue=total_yield_kg*price_per_kg
print("total revenue(INR):",total_revenue)

profit=total_revenue-total_input_cost
print("netprofit/loss(INR):",profit)

profit_per_kg=profit/total_yield_kg
print("profit per kg(INR):",profit_per_kg)

total_seed_kg=float(input("\nenter total seed available(kg):"))
seed_bag_capacity=float(input("enter seed bag capacity(kg):"))
full_seed_bags=total_seed_kg//seed_bag_capacity
remaining_seed=total_seed_kg%seed_bag_capacity
print("full seed bsg used:",full_seed_bags)
print("remaining seed (kg):",remaining_seed)

initial_crop_units=float(input("\nenter initial crop units:"))
growth_rate=float(input("enter seasonal growth rate (eg:1.10 for 10%):"))
seasons=int(input("enter number of seasons:"))
expected_crop=initial_crop_units*(growth_rate**seasons)
print("expected crops after seasons:",expected_crop)

weather_loss=float(input("enter loss due to weather(INR,enter positive value):"))
weather_loss=-weather_loss
adjusted_profit=profit+weather_loss


print("adjusted profit after weather impact(INR):",adjusted_profit)
positive_view=+adjusted_profit
negative_view=-adjusted_profit

print("positive profit view:",positive_view)
print("negative profit view:",negative_view)
print("\n=====end of farmer analysis=====")
```
