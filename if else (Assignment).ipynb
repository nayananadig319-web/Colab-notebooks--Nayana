```python
print("\n====CROP SUITABILITY CHECK====")
soil_moisture=50
rainfall=100
if soil_moisture>=70 and rainfall>=200:
     print("Suitable for rice")
else:
     print("Not suitable for rice")
```

```python
print("FERTILIZER RECOMMENDATION SYSTEM")
crop_name=(input("Crop name: "))
nitrogen_level=float(input("Enter Nitrogen level: "))
if crop_name=="rice" and nitrogen_level<50:
  print("Apply urea")
elif crop_name=="wheat" and nitrogen_level<40:
  print("Apply DAP")
else:
  print("Fertilizer not required")
```

```python
print("===CROP YIELD CLASSIFICATION===")
crop_yield=float(input("Enter crop yield in quintals: "))
if crop_yield>=50:
 print("High yield")
elif crop_yield>=30 and crop_yield<=49:
 print("Medium yield")
elif crop_yield<30:
 print("Low yield")
```
