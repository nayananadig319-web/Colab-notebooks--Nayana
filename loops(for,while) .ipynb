```python
days = [5,3,7,2]
windows = []
i = 0
while i < len(days)-1:
    j = i + 1
    while j < len(days):
        if days[i] + days[j] <= 8:
            windows.append((i,j))
        j += 1
    i += 1
print(windows)
```
