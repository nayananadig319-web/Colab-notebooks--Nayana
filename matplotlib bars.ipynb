```python
import matplotlib.pyplot as plt
data=[
    30,35,40,42,45,47,48,49,
    50,50,51,51,53,55,57,60,
    62,65,72,75
]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5, color="yellow",
    edgecolor="black",
    alpha=1.0,density=True,
    range=(30,80)
)
plt.xlabel("value")
plt.ylabel("Density")
plt.title("Histogram Without Random Data")
plt.grid(True)
plt.tight_layout()
plt.show()
```
