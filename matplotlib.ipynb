```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,25,20,30,40]
plt.plot(x,y,marker='o',linestyle='--',color='purple',linewidth=2,markersize=6,label='sales')
plt.xlabel("months")
plt.ylabel("sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y1=[10,18,25,30]
y2=[40,54,65,69]
plt.plot(x,y1,label='Line1',marker='o',linestyle='--',color='#800000')
plt.plot(x,y2,label='Line2',marker='*',linestyle=':',color='red')
plt.legend()
plt.twinx() #create second set of axes(shares the same x-axis but has independent y-axis)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(50)
y=np.random.rand(50)
sizes=1000*np.random.rand(50)
colors=np.random.rand(50)

plt.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.6)
plt.colorbar()
plt.title("Advanced Scatter Plot")
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```
