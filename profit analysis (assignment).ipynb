```python
def load_data(filename):
  crops=[]
  with open(filename,"r") as file:
    for line in file:
       line=line.strip()
       if line:
         data=line.split(",")
         crop={
            "name":data[0],
            "season":data[1],
            "area":int(data[2]),
            "yield":float(data[3]),
            "cost":float(data[4]),
            "revenue":int(data[5]),
            }
         crops.append(crop)
  return crops

def display_kharif_crops(all_crops):
  """Identifies and returns all kharif crops from a list of crops."""
  kharif_crops = [] # Initialize as a list
  for crop in all_crops: # Iterate over the passed list of all_crops
    if crop.get("season") == "Kharif": # Check 'season' key and value
      kharif_crops.append(crop.get("name")) # Append crop name
  return kharif_crops

def write_report(filename, crops, summary_data):
  file=open(filename,"w")
  file.write("Kharif Crops Analysis:\n")
  if summary_data: # Ensure summary_data exists before writing
      file.write(f"Found Kharif Crops: {', '.join(summary_data)}\n")
  file.close()

#-------main program------

def main():
  input_file="crop analysis.txt"
  output_file="profit_analysis.txt"
  crops=load_data(input_file)

  for crop in crops:
    print(crop)
  for crop in crops:
    for key,value in crop.items():
      print(f"{key}:{value}")
    print("-----")

  kharif_summary = display_kharif_crops(crops) # Pass the full list of crops
  write_report(output_file, crops, kharif_summary) # Pass the new summary variable
  print("\nanalysis complete")

main()
```
