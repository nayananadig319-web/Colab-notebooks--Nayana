```python
def load_data(filename):
  crops=[]
  file=open(filename,"r")
  for line in file:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={
          "name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "yield":float(data[3]),
          "cost":float(data[4]),
          "revenue":int(data[5])
      }
      crops.append(crop)
  file.close()
  return crops

def display_crops(crops):
  print("\n crop details:\n")
  for crop in crops:
      print (crop)

def display_kharif_crops(crops):
  print("\n kharif crops")
  for crop in crops:
    if crop["season"]=="kharif":
       print(crop["name"])

def display_non_kharif_crops(crops):
  print("\n non kharif crops")
  for crop in crops:
    if crop["season"]!="rabi":
      print(crop["name"])

def calculate_profit(crops):
  print("\n profit details:")
  for crop in crops:
    profit=crop["revenue"]-crop["cost"]
    print(crop["name"],"profit:",profit)

def summary_statistics(crops):
  total_area=0
  total_yield=0

  for crop in crops:

    total_area += crop["area"]
    total_yield += crop["yield"]

    avg_yield=total_yield/len(crops)

  print("summary statistics")
  print("total area:",total_area)
  print("average yield:",avg_yield)

#------main program-----

def main():
  input_file="data set (1).txt"
  output_file="profit analysis report.txt"

  crops=load_data(input_file)
  display_crops(crops)
  print("\n")

  for crop in crops:
    for key,value in crop.items():
     print(f"{key}:{value}")
    print("------")


  display_kharif_crops(crops)
  display_non_kharif_crops(crops)
  calculate_profit(crops)
  print("\n")
  summary_statistics(crops)

main()


```
