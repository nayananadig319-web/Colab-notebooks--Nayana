```python
def expenses_made():
  daily_expenses_list = []
  number_of_days = int(input("enter the number of days"))
  for i in range(number_of_days):
    day_name = input(f"enter the day name:")
    seed_cost = int(input("enter seed cost:"))
    fertilizer_cost = int(input("enter fertilizer cost:"))
    labour_cost = int(input("enter labour cost:"))

    daily_expenses_list.append([day_name, seed_cost, fertilizer_cost, labour_cost])
  return daily_expenses_list

# Call the function to get the daily expenses data
farm_daily_expenses = expenses_made()

overall_farm_expense = 0 # Initialize the overall expense variable
print("\n-----DAILY EXPENSE DETAILS-----")
for day_data in farm_daily_expenses:
  # day_data is [day_name, seed_cost, fertilizer_cost, labour_cost]
  day_name = day_data[0]
  seed_cost = day_data[1]
  fertilizer_cost = day_data[2]
  labour_cost = day_data[3]

  day_total = seed_cost + fertilizer_cost + labour_cost
  overall_farm_expense += day_total
  print(f"{day_name} Total expenses are: {day_total}")

print(f"\nOverall farm expense: {overall_farm_expense}")

crop_number=int(input("\nEnter number of crops: "))
crops=[]
```
